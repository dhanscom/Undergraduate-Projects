```python
import numpy as np
import pandas as pd
```

```python
# Read Data
data  = pd.read_csv('AMZN.csv')
# Remove Volume and Adjacent Close
data = data.drop(columns = ['Volume', 'Adj Close'])
# Print Data
print(data)
```

```python
# Find gap
def FindGap(data):
  gapList = []
  # Iterate through rows to find gaps
  for index, row in data.iterrows():
    if index + 1 < len(data):
      gapEnd = data.iloc[index+1]
      # Look for gaps up
      if row[2] < gapEnd[3]:
        add = (row[0],gapEnd[0], index+1,row[2],gapEnd[3], gapEnd[3]-row[2], row[2])
        gapList.append(add)
      # Look for gaps down
      if row[3] > gapEnd[2]:
        add = (row[0],gapEnd[0], index+1,row[3],gapEnd[2], gapEnd[2]-row[3], row[3])
        gapList.append(add)

  return gapList
```

```python
gaps = FindGap(data)
print(len(gaps))
```

```python
for i in gaps:
  print(i)
```

```python
# Checks for open gaps
def CheckGap(data, gaps):
  Open = []
  for gap in gaps:
    # Finds highest high and lowest low for dates after the gap is created
    SmallestLow = min(data.iloc[gap[2]+1:]['Low'])
    LargestHigh = max(data.iloc[gap[2]+1:]['High'])
    # If gap up and lowest low is greater than closing requirements then the gap is still open
    if gap[5] > 0 and SmallestLow > gap[6]:
      Open.append(gap)
    # If gap down and highest high is smaller than closing requirements then the gap is still open
    elif gap[5] < 0 and LargestHigh < gap[6]:
      Open.append(gap)
  return Open
```

```python
OpenGaps = CheckGap(data,gaps)
for i in OpenGaps:
  print(i)
```
